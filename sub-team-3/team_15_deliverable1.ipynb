```python
df.columns = df.columns.str.strip()
```

```python

text_cols = df.select_dtypes(include='object').columns

text_cols
```

```python
for col in text_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False).head(15))
```

```python
# Removing leading and trailing spaces from text values
for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
date_cols = [
    'Learner SignUp DateTime',
    'Opportunity End Date',
    'Date of Birth',
    'Entry created at',
    'Apply Date',
    'Opportunity Start Date'
]

for col in date_cols:
    print(f"\n--- {col} ---")
    print(df[col].dropna().head(10))
```

```python
for col in date_cols:
    converted = pd.to_datetime(df[col], errors='coerce')

    invalid = df[col].notna() & converted.isna()

    print(f"\n--- {col} ---")
    print("Invalid values:", invalid.sum())

    if invalid.sum() > 0:
        print(df.loc[invalid, col].head(10))
```

```python
for col in date_cols:
    converted = pd.to_datetime(df[col], errors='coerce')
    invalid = df[col].notna() & converted.isna()

    if invalid.sum() > 0:
        print(f"\n--- {col}: {invalid.sum()} invalid values ---")
        print(df.loc[invalid, col].value_counts().head(20))
```

```python
# Create backup copies before modifying anything
for col in date_cols:
    df[col + '_original'] = df[col]

# Extract only the date portion and convert it safely
for col in date_cols:
    df[col] = pd.to_datetime(
        df[col].astype('string').str.split().str[0],
        format='%m/%d/%Y',
        errors='coerce'
    )
```

```python
for col in date_cols:
    print(col, "-> Missing after cleaning:", df[col].isna().sum())
```

```python
df.loc[df['Apply Date'].isna(), ['Apply Date', 'Apply Date_original']]
```

```python
mask = df['Apply Date'].isna() & df['Apply Date_original'].notna()

df.loc[mask, 'Apply Date'] = pd.to_datetime(
    df.loc[mask, 'Apply Date_original'].str.split().str[0],
    format='%d/%m/%Y',
    errors='coerce'
)
```

```python
df['Apply Date'].isna().sum()
```

```python
df[df['Institution Name'].isna()][
    ['Institution Name', 'Current/Intended Major']
]
```

```python
df['Institution Name'] = df['Institution Name'].fillna('Unknown')
df['Current/Intended Major'] = df['Current/Intended Major'].fillna('Unknown')
```

```python
print("Institution Name:", df['Institution Name'].isna().sum())
print("Current/Intended Major:", df['Current/Intended Major'].isna().sum())
```

```python
categorical_cols = [
    'Gender',
    'Country',
    'Opportunity Category',
    'Status Description'
]

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False))
```

```python
print("Dataset Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)
```

```python
original_cols = [col for col in df.columns if col.endswith('_original')]
df.drop(columns=original_cols, inplace=True)

print("Final shape:", df.shape)
```

```python
df.to_csv("Cleaned_RIT_Opportunity_Data.csv", index=False)
```
